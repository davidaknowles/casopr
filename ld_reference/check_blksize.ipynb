```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
blk = pd.read_csv('/gpfs/commons/groups/knowles_lab/data/ADSP_reguloML/LD/LD_ADSP36K_4PRScs_OCT/snp_count.txt', sep =' ', names=['chr','blk','start','end','snp_count'])
```

```python
blk['chr'] = blk.chr.str.replace(r'chr', '', regex=True).astype(int)
```

```python
blk.plot("chr", "snp_count", linestyle='None', marker='x', markersize=5, color='dodgerblue', figsize=(20,8))
plt.xticks(blk.chr, fontsize=10)
plt.ylabel('SNP count per LD block', fontsize=14)
plt.axhline(y=4000, color='r', linestyle='--', linewidth=2)


for i, row in blk.iterrows():
    if row['snp_count'] > 4000:
        plt.annotate(f"{row['blk']}", (row['chr'], row['snp_count']),
                     textcoords="offset points", xytext=(9.5, 0), ha='center', fontsize=9, color='darkblue')
plt.show()
```
