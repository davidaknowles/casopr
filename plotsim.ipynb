```python
# import seaborn as sns
# import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

# Assuming 'betas' is your DataFrame
columns_to_plot = ['beta_marginal', 'beta_est1', 'beta_prscs']
subset_betas = betas[columns_to_plot]

# Melt the DataFrame to long format for the violin plot
melted_betas = subset_betas.melt(var_name='Variable', value_name='Value')

# Perform Mann-Whitney U test for each pair
pairs = [('beta_marginal', 'beta_est1'), ('beta_marginal', 'beta_prscs'), ('beta_prscs', 'beta_est1')]
p_values = []

for pair in pairs:
    variable1, variable2 = pair
    _, p_value = mannwhitneyu(melted_betas[melted_betas['Variable'] == variable1]['Value'],
                              melted_betas[melted_betas['Variable'] == variable2]['Value'])
    p_values.append(p_value)

# Set a larger font size
sns.set(font_scale=1.5)
sns.set_style("whitegrid")

# Create a violin plot using Seaborn
plt.figure(figsize=(12, 8))
ax = sns.violinplot(x='Variable', y='Value', data=melted_betas)

#Annotate p-values on the plot
for i, pair in enumerate(pairs):
    variable1, variable2 = pair
    p_value = p_values[i]
    y_position = melted_betas['Value'].max() + i  # Adjust y-position for annotation
    ax.text(i, y_position, f'p-value = {p_value:.4f}', ha='center', va='center', color='r', fontsize=14)

# plt.title('Violin Plot of Beta Values with P-Values', fontsize=20)
plt.xlabel('')
plt.ylabel('Beta', fontsize=16)
plt.show()
```
